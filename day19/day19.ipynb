### Part One

As your [probe](https://adventofcode.com/2021/day/17) drifted down through this area, it released an assortment of _beacons_ and _scanners_ into the water. It's difficult to navigate in the pitch black open waters of the ocean trench, but if you can build a map of the trench using data from the scanners, you should be able to safely reach the bottom.

The beacons and scanners float motionless in the water; they're designed to maintain the same position for long periods of time. Each scanner is capable of detecting all beacons in a large cube centered on the scanner; beacons that are at most 1000 units away from the scanner in each of the three axes (`x`, `y`, and `z`) have their precise position determined relative to the scanner. However, scanners cannot detect other scanners. The submarine has automatically summarized the relative positions of beacons detected by each scanner (your puzzle input).

For example, if a scanner is at `x,y,z` coordinates `500,0,-500` and there are beacons at `-500,1000,-1500` and `1501,0,-500`, the scanner could report that the first beacon is at `-1000,1000,-1000` (relative to the scanner) but would not detect the second beacon at all.

Unfortunately, while each scanner can report the positions of all detected beacons relative to itself, _the scanners do not know their own position_. You'll need to determine the positions of the beacons and scanners yourself.

The scanners and beacons map a single contiguous 3d region. This region can be reconstructed by finding pairs of scanners that have overlapping detection regions such that there are _at least 12 beacons_ that both scanners detect within the overlap. By establishing 12 common beacons, you can precisely determine where the scanners are relative to each other, allowing you to reconstruct the beacon map one scanner at a time.

Unfortunately, there's a second problem: the scanners also don't know their _rotation or facing direction_. Due to magnetic alignment, each scanner is rotated some integer number of 90-degree turns around all of the `x`, `y`, and `z` axes. That is, one scanner might call a direction positive `x`, while another scanner might call that direction negative `y`. Or, two scanners might agree on which direction is positive `x`, but one scanner might be upside-down from the perspective of the other scanner. In total, each scanner could be in any of 24 different orientations: facing positive or negative `x`, `y`, or `z`, and considering any of four directions "up" from that facing.

### Part Two

Sometimes, it's a good idea to appreciate just how <span title="The deepest parts of the ocean are about as deep as the altitude of a normal commercial aircraft, roughly 11 kilometers or 36000 feet.">big</span> the ocean is. Using the [Manhattan distance](https://en.wikipedia.org/wiki/Taxicab_geometry), how far apart do the scanners get?

In the above example, scanners `2` (`1105,-1205,1229`) and `3` (`-92,-2380,-20`) are the largest Manhattan distance apart. In total, they are `1197 + 1175 + 1249 = 3621` units apart.

_What is the largest Manhattan distance between any two scanners?_


```python
import numpy as np
import copy
from time import perf_counter
from collections import Counter, deque
from functools import lru_cache
import numba
from itertools import permutations, combinations, product
```

```python
f = open("input.txt", "r")
rawstring = f.read()
f.close()
```

```python
scanners = []
lines = rawstring.splitlines()
temp = []
for line in lines:
    if 'scanner' in line:
        if len(temp) != 0:
            scanners.append(np.array(temp))
        temp = []
    elif len(line) != 0:
        temp.append([int(x) for x in line.split(",")])
scanners.append(np.array(temp))
```

```python
pairwise_distances = []
for i in range(len(scanners)):
    temp = np.zeros((scanners[i].shape[0],scanners[i].shape[0]),dtype='i,i,i')
    for l, j in combinations(range(scanners[i].shape[0]),2):
        #temp[l,j] = (np.abs(scanners[i][l][0] - scanners[i][j][0])**2+np.abs(scanners[i][l][1] - scanners[i][j][1])**2+np.abs(scanners[i][l][2] - scanners[i][j][2])**2)**0.5
        _ = (
            np.abs(scanners[i][l][0] - scanners[i][j][0]),
            np.abs(scanners[i][l][1] - scanners[i][j][1]),
            np.abs(scanners[i][l][2] - scanners[i][j][2]))
        _ = tuple(sorted(list(_)))
        temp[l,j] = _
    pairwise_distances.append(temp)
```

```python
def index2pair(obsIndex, flatIndex):
    edge_len = pairwise_distances[obsIndex].shape[0]
    x = flatIndex//edge_len
    y = flatIndex%edge_len
    return x,y
```

```python
overlapping = np.zeros((len(scanners),len(scanners)))
```

```python
confirmed = dict()
for x1, x2 in combinations(range(len(scanners)),2):
    d,i,c = np.unique(pairwise_distances[x1],return_index=True,return_counts=True)
    d = d[c == 1]
    i = i[c == 1]
    c = c[c == 1]
    d2,i2,c2 = np.unique(pairwise_distances[x2],return_index=True,return_counts=True)
    d2 = d2[c2 == 1]
    i2 = i2[c2 == 1]
    c2 = c2[c2 == 1]
    mapping = dict()
    for val in np.intersect1d(d,d2):
        origin = index2pair(x1,i[np.where(val == d)])
        destination = index2pair(x2,i2[np.where(val == d2)])
        for key in origin:
            if int(key) not in mapping.keys():
                mapping[int(key)] = set(list(np.array(destination).flat))
            else:
                mapping[int(key)] = mapping[int(key)].intersection(set(list(np.array(destination).flat)))
    overlapping[x1,x2] = sum([1 for x in mapping.values() if len(x) == 1])
    for key, value in [(key, value) for key, value in mapping.items() if len(value) == 1]:
        confirmed[(x1, key)] = (x2, list(value)[0])
    
```

```python
confirmedl = list()
for k,v in confirmed.items():
    confirmedl.append(set([k]+[v]))
```

```python
for _ in range(len(scanners)):
    for i in range(len(confirmedl)):
        for y in range(len(confirmedl)):
            if y != i:
                if confirmedl[i].intersection(confirmedl[y]) != set():
                    confirmedl[i] = confirmedl[i].union(confirmedl[y])
```

```python
mul = list(dict.fromkeys([tuple(sorted(list(tuple(x)))) for x in confirmedl]))
```

```python
mul2 = set()
count = 0
for l in mul:
    count += 1
    for t in l:
        mul2.add(t)
```

```python
for i in range(len(scanners)):
    for j in range(len(scanners[i])):
        if tuple((i, j)) not in mul2:
            count+=1
            mul2.add(tuple((i, j)))
print(count)
```

have to fix the coordinate systems so they show the same relative distances between two points:

```python
known = set([0])
unknown = set(list(range(len(scanners)))).difference(known)
scannerLocs = [np.array([0,0,0])]

def find_possible_points():
    u_p = []
    k_p = []
    for specific_known in known:
        for specific_unkown in unknown:
            u_p.clear()
            k_p.clear()
            for point in mul:
                for x,y in point:
                    if specific_known == x and specific_unkown in [x for x,y in point]:
                        k_p.append((x,y))
                        if len(k_p)+len(u_p) == 4:
                            return k_p + u_p, specific_known, specific_unkown
                    if specific_unkown == x and specific_known in [x for x,y in point]:
                        u_p.append((x,y))
                        if len(k_p)+len(u_p) == 4:
                            return k_p + u_p, specific_known, specific_unkown

for i in range(1, len(scanners)):
    # need to find points which appear in somewhere in our known point cloud 
    # and in the unknown one:
    possible_points, k, u = find_possible_points()
    p1_k = possible_points[0][1]
    p2_k = possible_points[1][1]
    p1_u = possible_points[2][1]
    p2_u = possible_points[3][1]

    # swapping axis' until the magnitudes are correct
    for t in permutations(range(3),3):
        temp = scanners[u][:,list(t)]
        ref = scanners[k][p1_k]-scanners[k][p2_k]
        y = temp[p1_u]-temp[p2_u]
        if(np.allclose(np.abs(ref),np.abs(y))):
            # committing to change
            scanners[u] = scanners[u][:,list(t)]
            for sign in list(product([1,-1],repeat=3)):
                temp = scanners[u]
                y = (temp[p1_u]-temp[p2_u])*sign
                if(np.allclose(ref,y)):
                    # commit to change
                    scanners[u] = scanners[u]*sign
                    known.add(u)
                    unknown = unknown.difference(known)
                    scannerPos = scanners[k][p1_k] - scanners[u][p1_u]
                    scanners[u] += scannerPos
                    scannerLocs.append(scannerPos)
                    break
```
